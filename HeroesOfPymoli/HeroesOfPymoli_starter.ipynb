```python

import pandas as pd

file_to_load = "Resources/purchase_data.csv"

purchase_data = pd.read_csv(file_to_load)
purchase_data.head(10)
```

## Player Count

* Display the total number of players


```python
players = len(purchase_data["SN"].unique())
players_df = pd.DataFrame({"Total Players":[players]})
players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
items_unique = len(purchase_data["Item ID"].unique())

average_purchase_price = purchase_data["Price"].mean()

Total_purchases = purchase_data["Purchase ID"].count()

total_revenue = purchase_data["Price"].sum()

purchase_summary_df = pd.DataFrame({"NUMBER OF UNIQUE ITEMS": items_unique,
                                      "AVERAGE PRICE": [average_purchase_price],
                                      "NUMBER OF PURCHASES": [Total_purchases],
                                        "TOTAL REVENUE": [total_revenue]})

purchase_summary_df["AVERAGE PRICE"] = purchase_summary_df["AVERAGE PRICE"].map("{:.2f}".format)

purchase_summary_df["TOTAL REVENUE"] = purchase_summary_df["TOTAL REVENUE"].map("{:}".format)

purchase_summary_df

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender = purchase_data.drop_duplicates('SN')

gender_count = gender["Gender"].value_counts()

gender_percent = (gender['Gender'].value_counts()/players)*100

gender_df = pd.DataFrame({
        'Count': gender_count,
        'Percent of Total Players': gender_percent.map('{:.2f}%'.format)
})
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchase_df = purchase_data.groupby(["Gender"])

gender_purchase_count = gender_purchase_df["Purchase ID"].count()
gender_purchase_ave = gender_purchase_df["Price"].mean().map("${:.2f}".format)
gender_purchase_total = gender_purchase_df["Price"].sum().map("${:,.2f}".format)

gender_purchase_ave_person = (gender_purchase_df["Price"].sum() / 
                                gender_purchase_df["SN"].count()).map("${:.2f}".format)

gender_purchase_analysis_df = pd.DataFrame({"Purchase Count":gender_purchase_count,
                                       "Average Purchase Price":gender_purchase_ave,
                                       "Total Purchase Value":gender_purchase_total,
                                       "Avg Total Purchase per Person":gender_purchase_ave_person})
gender_purchase_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 400]

age_bins = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["age_group"] =  pd.cut(purchase_data["Age"], bins, labels=age_bins)
players_age = purchase_data.groupby("age_group").SN.nunique()  

age_demography = pd.DataFrame({
    "Total Count": players_age,
    "Percentage of Players": (players_age / purchase_data.SN.nunique() * 100).map("{:.2f}".format)
})
age_demography
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_age_count_df = purchase_data[['age_group', 'Purchase ID']]
purchase_price_aga_df = purchase_data[['age_group', 'Price']]

total_purchase_per_person = purchase_age_count_df.groupby('age_group').count() 

total_purchase_values_age = purchase_price_aga_df.groupby('age_group').sum()

purchase_analysis_age = pd.DataFrame({
    'Purchase Count': total_purchase_per_person['Purchase ID'],
    'Average Purchase Price': (total_purchase_values_age['Price'] / total_purchase_per_person['Purchase ID']).map("${:.2f}".format),
    'Total Purchase Value': total_purchase_values_age['Price'].map("${:.2f}".format)
})
purchase_analysis_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders_purchase_df = purchase_data.groupby(["SN"])

spender_purchase_Count = spenders_purchase_df["Purchase ID"].count()
spenders_ave_purchase_price = spenders_purchase_df["Price"].mean()
total_spenders_purchase_value = spenders_purchase_df["Price"].sum()

spenders_summary_df = pd.DataFrame({"Purchase Count": spender_purchase_Count,
                                    "Average Purchase Price": spenders_ave_purchase_price,
                                    "Total Purchase Value": total_spenders_purchase_value})

spenders_summary_df = spenders_summary_df.sort_values(["Total Purchase Value"], ascending=False)

spenders_summary_df["Average Purchase Price"] = spenders_summary_df["Average Purchase Price"].map("${:,.2f}".format)
spenders_summary_df["Total Purchase Value"] = spenders_summary_df["Total Purchase Value"].map("${:,.2f}".format)


spenders_summary_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items_df = purchase_data[["Item ID", "Item Name", "Price"]]

items_group_df = popular_items_df.groupby(["Item ID", "Item Name"])

item_count = items_group_df["Item ID"].count()
item_price = items_group_df["Price"].mean()
total_item_vol = items_group_df["Price"].sum()

popular_items_df = pd.DataFrame({"Purchase Count": item_count,
                                 "Item Price": item_price,
                                 "Total Purchase Value": total_item_vol})

popular_items_df = popular_items_df.sort_values(["Purchase Count"], ascending=False)

popular_items_df["Item Price"] = popular_items_df["Item Price"].map("${:,.2f}".format)
popular_items_df["Total Purchase Value"] = popular_items_df["Total Purchase Value"].map("${:,.2f}".format)

popular_items_df
```
